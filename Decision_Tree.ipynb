# Predicting whose Salary will be more than 100K

```python
import pandas as pd
```

```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("salaries.csv")
df
```

```python
X=df.drop("salary_more_then_100k",axis="columns")
X
```

```python
y=df["salary_more_then_100k"]
y
```

```python
from sklearn.preprocessing import LabelEncoder
v_company=LabelEncoder()
v_job=LabelEncoder()
v_degree = LabelEncoder()
```

```python
X["company_n"]=v_company.fit_transform(X["company"])
X["job_n"]=v_job.fit_transform(X["job"])
X["degree_n"]=v_degree.fit_transform(X["degree"])
```

```python
X
```

```python
X=X.drop(["company","job","degree"],axis="columns")
```

```python
X
```

```python
y
```

```python
from sklearn import tree
model=tree.DecisionTreeClassifier()
```

```python
model.fit(X,y)
```

```python
model.predict([[2,1,0]])
```

```python
model.score(X,y)
```

```python
prediction = model.predict([[2,0,0]])
```

```python
if (prediction[0]==1):
    print("salary is more than 100k")
else:
    print("salary is less then 100k")
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=10)
```

```python
X_train
```

```python
model.fit(X_train,y_train)
```

```python
y_predict=model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
modelacc=accuracy_score(y_predict,y_test)
modelacc
```

# titanic survival prediction

```python
df=pd.read_csv("titanic.csv")
df
```

```python
X=df[["Pclass","Sex","Age","Fare"]]
X
```

```python
y=df["Survived"]
y
```

```python
X.Sex=X.Sex.replace({"male":1,"female":2})
X
```

```python
X["Age"]=X["Age"].fillna(X["Age"].mean())
```

```python
from sklearn import tree
model=tree.DecisionTreeClassifier()
```

```python
model.fit(X,y)
```

```python
model.score(X,y)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=10)
```

```python
model.fit(X_train,y_train)
```

```python
y_predict=model.predict(X_test)
y_predict
```

```python
y_test
```

```python
model.score(X_test,y_test)
```

```python
modelacc=accuracy_score(y_predict,y_test)
modelacc
```
